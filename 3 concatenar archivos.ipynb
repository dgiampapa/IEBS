# Concatenar archivos de prueba

Estos estan elaborados con la libreria de analisis de sentimiento

```python
import pandas as pd

ruta1 = '/content/drive/MyDrive/IEBS/Global Proyect/datos/scoreprueba/score-1372470686.csv'
ruta2 = '/content/drive/MyDrive/IEBS/Global Proyect/datos/scoreprueba/score-230123496.csv'
ruta3 = '/content/drive/MyDrive/IEBS/Global Proyect/datos/scoreprueba/score-270429778.csv'
ruta4 = '/content/drive/MyDrive/IEBS/Global Proyect/datos/scoreprueba/score-86031609.csv'
ruta5 = '/content/drive/MyDrive/IEBS/Global Proyect/datos/scoreprueba/score-9980072.csv'

df1 = pd.read_csv(ruta1)
df2 = pd.read_csv(ruta2)
df3 = pd.read_csv(ruta3)
df4 = pd.read_csv(ruta4)
df5 = pd.read_csv(ruta5)
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
print(df4.shape)
print(df5.shape)
```

```python
dft = pd.concat([df1,df2,df3,df4,df5])
```

```python
dft.shape
```

```python
ruta_grabar = '/content/drive/MyDrive/IEBS/Global Proyect/datos/scoreprueba/score-total.csv'

dft.to_csv(ruta_grabar)
```
